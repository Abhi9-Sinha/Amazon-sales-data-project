```python
import matplotlib.pyplot as plt

# First, convert the 'Date' column to a datetime object if it's not already in that format
df['Date'] = pd.to_datetime(df['Date'])

# Extract the day of the month (DD format) from the 'Date' column
df['Day'] = df['Date'].dt.day

# Count the frequency of each day
day_counts = df['Day'].value_counts().sort_index()

# Set the figure size (optional)
plt.figure(figsize=(12, 6))

# Create a bar chart for day frequency
day_counts.plot(kind='bar', color='blue')
plt.title('Day Frequency')
plt.xlabel('Day of the Month')
plt.ylabel('Frequency')

# Show the bar chart
plt.show()
```

```python
# STATUS
```

```python
import matplotlib.pyplot as plt

# Calculate the value counts and percentages
status_counts = df['Status'].value_counts()
percentages = status_counts / status_counts.sum() * 100

# Create the pie chart
colors = plt.cm.Paired(range(len(status_counts)))  # Generate a list of colors

plt.figure(figsize=(8, 8))  # Set the figure size (optional)

# Create the pie chart without percentages displayed on the chart
wedges, texts = plt.pie(
    status_counts,
    
    startangle=140,
    colors=colors
)

# Create a legend for category names separately
plt.legend(status_counts.index, title="Status", loc="upper left", bbox_to_anchor=(1, 1))

# Create a separate legend for percentage distribution
legend_labels = [f"{label}: {percent:.1f}%" for label, percent in zip(status_counts.index, percentages)]
plt.legend(legend_labels, title="Percentage", loc="lower left", bbox_to_anchor=(1, 0))

plt.title('Status Distribution')

# Display the chart
plt.show()
```

```python
#AMOUNT
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate statistics for the 'Amount' column
amount_stats = df['Amount'].describe()

# Exclude the 'count' statistic
amount_stats = amount_stats.drop('count')

# Add median to the statistics table
amount_stats['Median'] = df['Amount'].median()

# Create a custom statistics DataFrame
statistics_df = pd.DataFrame({
    'Statistic': amount_stats.index,
    'Value': amount_stats.values
})

# Plot the statistics
plt.figure(figsize=(10, 6))
plt.bar(statistics_df['Statistic'], statistics_df['Value'], color=['blue', 'green', 'red', 'purple', 'orange'])
plt.title('Statistics for Amount Column')
plt.xlabel('Statistic')
plt.ylabel('Value')

# Add text labels for each bar
for i, row in statistics_df.iterrows():
    plt.text(i, row['Value'], f'{row["Value"]:.2f}', ha='center', va='bottom')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each state
state_counts = df['ship-state'].value_counts()

# Remove states with no frequency (count > 0)
state_counts = state_counts[state_counts > 0]

# Optionally, you can remove outliers based on a threshold count
# Replace 'threshold_count' with your desired threshold
threshold_count = 10  # Example threshold, adjust as needed
state_counts = state_counts[state_counts >= threshold_count]

# Sort the states by frequency (optional)
state_counts = state_counts.sort_values(ascending=False)

# Set the figure size (optional)
plt.figure(figsize=(12, 6))

# Create the bar chart
plt.bar(state_counts.index, state_counts.values, color='green')

# Add labels and title
plt.title('Ship State Frequency')
plt.xlabel('State')
plt.ylabel('Frequency')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=90)

# Show the bar chart
plt.show()
```
